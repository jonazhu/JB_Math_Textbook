3.1.1.
We know that
 $\frac{dx}{dt}  =  v $
 and $v = \frac{p}{m}$.
 So, $ \frac{dx}{dt} = \frac{\partial H}{\partial p}$ since the partial derivative of $H$ with respect to $p$ is $\frac{p}{m}$.

We also know that $\frac{dp}{dt} = F$ and $F = -kx$. So, $\frac{dp}{dt} = \frac{- \partial H}{\partial x}$ since the partial derivative of $H$ with respect to x is $kx$.

3.1.2.
$\frac{dE}{dt} = (\frac{ \partial H}{\partial x} * \frac{dx}{dt}) + (\frac{\partial H}{\partial p} * \frac{dp}{dt}) = (\frac{-dp}{dt} * \frac{dx}{dt}) + (\frac{dx}{dt} * \frac{dp}{dt}) = -FV + VF = 0$.

 Taking the partial derivative of the function $f_T(m)$ and setting it equal to zero, we obtain
$
\frac{\partial}{\partial m} = m - \frac{m}{t} + \frac{m^3}{3} = 0
$

m = 0

$1 - \frac{1}{T} + \frac{m^2}{3} = 0 \rightarrow
\frac{m^2}{3} = \frac{1-T}{T} \rightarrow
m^2 = \frac{3-3T}{T} \rightarrow
m = \pm \sqrt{\frac{3-3T}{T}}
$


Setting what we found for m equal to zero and solving for the critical T value we obtain
$
\sqrt{\frac{3}{T}-3} = 0 \rightarrow \frac{3}{T} = 3 \rightarrow 3T = 3 \rightarrow T_crit=1
$

Taking the second derivative of the original equation we obtain
$
f''(m) = 1 - \frac{1}{T} + m^2
$

Comparing values with $T_{crit}$ in the second derivative, if $T \geq 1$, $f''(0)>0$, If $T<1$, $f''(0) < 0$. 
It follows that for values of $T\geq1$, $m=0$ is the absolute minimum and for $T<1$, $m=\pm m_0$ is a symmetric absolute minimum
