# 8. Normal Distribution and Process Control

In this lab we will apply the Normal distribution to statistical process control. The main goal of the latter is to ensure the output of a manufacturing process remains within limits determined by usual random  variation. If a very low probability event occurs, this may be used as a signal that there is a problem with the manufacturing process, in which case the process might be stopped and equipment checked for possible defects.

<h4>Derivatives of Exponential Functions</h4>

The normal distribution is given by 

$$f(x)=\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}},$$

where $\mu$ is a constant called the <i>mean</i> and $\sigma$ is another constant called the <i>standard deviation.</i>  

To analyze the normal distribution using calculus, we need the derivative formula

$$\frac{d}{dx} e^{u(x)}= e^{u(x)}u'(x).$$

For example, let $f(x) = c_1 e^{-\frac{(x-c_2)^2}{2c_3^2}}$. Then,

$$f'(x)= c_1 e^{-\frac{(x-c_2)^2}{2c_3^2}} [\frac{-2(x-c_2)}{2c_3^2}]=c_1 e^{-\frac{(x-c_2)^2}{2c_3^2}} [\frac{-(x-c_2)}{c_3^2}]$$

Note that $f'(c_2)=0$.  Since $f'(x)>0$ ($f$ is increasing) for all $x<c_2$ and $f'(x)<0$ ($f$ is decreasing) for all $x>c_2$, we know $f(x)$ has an absolute maximum value at $x=c_2$.

We use the product rule to find $f''(x)$:

$$ f''(x)= c_1 e^{-\frac{(x-c_2)^2}{2c_3^2}} [\frac{-(x-c_2)}{c_3^2}]^2+c_1 e^{-\frac{(x-c_2)^2}{2c_3^2}}(-\frac{1}{c_3^2})$$

$$= c_1 e^{-\frac{(x-c_2)^2}{2c_3^2}}[\frac{(x-c_2)^2}{c_3^4}-\frac{1}{c_3^2}].$$

To find the $x$-coordinates of the two inflection points, set $f''(x)=0$:

$$\frac{(x-c_2)^2}{c_3^4}-\frac{1}{c_3^2}=0 \Rightarrow$$

$$\frac{(x-c_2)^2}{c_3^4}= \frac{1}{c_3^2}\Rightarrow $$

$$(x-c_2)^2 = c_3^2 \Rightarrow x-c_2= \pm c_3 \Rightarrow$$

$$x=c_2\pm c_3.$$

The inflection points occur symmetrically about the maximum point at $x=c_2$.

<h4> Rule of 68-95-99.7 </h4>

If data points are normally distributed

68% of the data will be within 1 standard deviation of the mean

95% of the data will be within 2 standard deviations of the mean

99.7% of the data will be within 3 standard deviations of the mean.

The plot below shows the standard normal distribution with mean $\mu=0$ and standard deviation $\sigma=1$. The total area under the curve is equal to 1.  The shaded area on the graph below represents the probability that standard normally distributed data will be within 1 standard deviation of the mean. 



````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-4,4,100)
f= (1/np.sqrt(2*np.pi))*np.exp(-x**2/2)  #standard (unit) normal distribution
plt.figure(figsize=(8, 4))
plt.plot(x,f,color='k',label="Normal Distribution (mu=0, sigma=1)")
plt.xlim((-4.,4))
plt.ylim((0,.5))
x1=np.linspace(-1,1,50)
y1= (1/np.sqrt(2*np.pi))*np.exp(-x1**2/2) 
plt.fill_between(x1,y1,color='lightgray')
plt.text(-.85,.2,'Shaded area = .68')
plt.grid()
plt.legend()
plt.show()
````

````python
x = np.linspace(-4,4,100)
f= (1/np.sqrt(2*np.pi))*np.exp(-x**2/2)  #standard (unit) normal distribution
plt.figure(figsize=(8, 4))
plt.plot(x,f,color='k',label="Normal Distribution (mu=0, sigma=1)")
plt.xlim((-4.,4))
plt.ylim((0,.5))
x2=np.linspace(-2,2,50)
y2= (1/np.sqrt(2*np.pi))*np.exp(-x2**2/2) 
plt.fill_between(x2,y2,color='lightblue')
plt.text(-.85,.2,'Shaded area = .95')
plt.grid()
plt.legend()
plt.show()
````

````python
x = np.linspace(-4,4,100)
f= (1/np.sqrt(2*np.pi))*np.exp(-x**2/2)  #standard (unit) normal distribution
plt.figure(figsize=(8, 4))
plt.plot(x,f,color='k',label="Normal Distribution (mu=0, sigma=1)")
plt.xlim((-4.,4))
plt.ylim((0,.5))
x3=np.linspace(-3,3,50)
y3= (1/np.sqrt(2*np.pi))*np.exp(-x3**2/2) 
plt.fill_between(x3,y3,color='lightgreen')
plt.text(-.85,.2,'Shaded area = .997')
plt.grid()
plt.legend()
plt.show()
````
