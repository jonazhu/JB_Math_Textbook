# 4 Matplotlib

Matplotlib was created by John Hunter to help make graphs.  Like numpy and pandas, matplotlib is a widely used Python library. Dr. Hunter developed this library as a neurobiologist, studying epilepsy at the University of Chicago. Unfortunately, Dr. Hunter died of cancer at age 44.

1) The matplotlib (mpl) library has a sub-library called pyplot (plt).  We can access the functions in pyplot by executing the next cell.

```python
import matplotlib.pyplot as plt
```

2) Here is an example of a simple plot.

```python
# create a new figure
plt.figure()
# plot the points (0, 1) (1,0),(2,1),(3,0) and (4,1) with a 'o' marker and connect them using  'o-' 
xvalues=[0,1,2,3,4]
yvalues=[1,0,1,0,1]
plt.plot(xvalues, yvalues, 'o-',color='blue')
```
