3. Plot the direction fields

    - $ \dfrac{dy}{dx}= x^2-y^2-y $

````python
import numpy as np
import matplotlib.pyplot as plt

nx, ny = .3, .3
x = np.arange(-2,2,nx)
y = np.arange(-2,2,ny)

#Create a rectangular grid with points
X,Y = np.meshgrid(x,y)

#Define the functions
dy = X**2 - Y**2 - Y
dx = np.ones(dy.shape)

#Plot
plt.quiver(X,Y,dx,dy,color='purple')
plt.title('Direction Field for dy/dx = x^2 - y^2 - y')
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

nx, ny = .3, .3
x = np.arange(-2,2,nx)
y = np.arange(-2,2,ny)

#Create a rectangular grid with points
X,Y = np.meshgrid(x,y)

#Define the functions
dy = Y*(X-Y)
dx = np.ones(dy.shape)

#Plot
plt.quiver(X,Y,dx,dy,color='purple')
plt.title('Direction Field for dy/dx = y(x - y)')
plt.show()
````
